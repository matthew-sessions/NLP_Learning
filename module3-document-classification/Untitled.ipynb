```python
import pandas as pd
from scipy import stats
```

```python
s1 = pd.read_csv('submission1.csv')
s2 = pd.read_csv('submission2.csv')
s3 = pd.read_csv('submission3.csv')
s4 = pd.read_csv('submission4.csv')
```

```python
lii = []
for i in range(len(s1)):
    one = s1.values[i][1]
    two = s2.values[i][1]
    three = s3.values[i][1]
    four = s4.values[i][1]
    at = stats.mode([one,two,three,four])
    lii.append(at[0][0])
```

```python
lii
```

```python
df = s1
df.head()
```

```python
df['category'] = lii
```

```python
df.to_csv('submission5.csv', index=False)
```
